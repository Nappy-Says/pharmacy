```python
#print(df.to_dict())

dic = []
for i in range(1,124):
    dic.append({
        "code" : df[0][i],
        "country" : df[1][i]
    })
```

```python
d = []
for i in dic:
    if '–' in i['code'] and not '&' in i['code']:
        #print(i)
        for j in range(  int(i['code'].split(' – ')[0]), int(i['code'].split(' – ')[1]) ):
            d.append({
                "code" : str(j),
                "country" : i['country']
            })
    elif '&' in i['code']:
        for k in i['code'].split(' & '):
            for j in range(  int(k.split(' – ')[0]), int(k.split(' – ')[1]) ):
                d.append({
                    "code" : j,
                    "country" : i['country']
                })            
    else:
        d.append({
           'code' : i['code'],
           'country' : i['country'] 
        })
```

```python
d
```

```python
endict = {}

for i in d:
    endict[i['code']] = i['country']

endict
```
